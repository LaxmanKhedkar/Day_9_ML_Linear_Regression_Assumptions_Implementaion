```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
df_cs.info()
```

```python
df_cs.columns
```

```python
considered_feature_01 = ['Sales_in_thousands','__year_resale_value','Engine_size', 'Horsepower',
       'Wheelbase', 'Width', 'Length', 'Curb_weight', 'Fuel_capacity',
       'Fuel_efficiency', 'Power_perf_factor']
```

```python
def compute_vif(features):
    x = df_cs[features]
    x["constant"] = 1
    vif = pd.DataFrame()
    vif["Variable"] = x.columns
    vif["VIF"] = [variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
    vif = vif[~(vif["Variable"]=="constant")]
    vif = vif.sort_values("VIF",ascending = False).reset_index(drop = True)
    return vif
```

```python
compute_vif(considered_feature_01)
```

```python
considered_feature_02 = ['Sales_in_thousands','__year_resale_value','Engine_size', 'Horsepower',
       'Wheelbase', 'Width', 'Length', 'Curb_weight', 'Fuel_capacity',
       'Fuel_efficiency']
```

```python
compute_vif(considered_feature_02)
```
